```python
def plot_corr(df, size=10):
    corr = df.corr()    
    fig, ax = plt.subplots(figsize = (size, size))
    ax.matshow(corr)  
    plt.xticks(range(len(corr.columns)), corr.columns) 
    plt.yticks(range(len(corr.columns)), corr.columns)  
```

```python
plot_corr(df)
```

```python
from sklearn import preprocessing
```

```python
df2=preprocessing.scale(df)
```

```python
from  sklearn.cluster import KMeans
```

```python
wcss = []
 
for i in range(1, 5):
    kmeans = KMeans(n_clusters = i, init = 'random')
    kmeans.fit(df2)
    print (i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 5), wcss)
plt.title('O Metodo Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS') #within cluster sum of squares
plt.show()
```

```python
kmeans=KMeans(n_clusters = 3, init='random')
```

```python
from sklearn.model_selection import train_test_split
df3=pd.read_csv("wine.csv")
Y=df3['Customer_Segment']
split_test_size = 0.30
X_treino, X_teste, Y_treino, Y_teste = train_test_split(df2, Y, test_size = split_test_size, random_state = 42)
```

```python
kmeans.fit(X_treino)
```

```python
labels = kmeans.labels_
labels
```

```python
plt.scatter(X_treino[:, 0], X_treino[:,1], s = 100, c = kmeans.labels_)
#plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.title('Iris Clusters and Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')

plt.show()
```

```python
from sklearn import metrics
from numpy import ndarray
```

```python
nb_predict = kmeans.labels_+1
for i in range(124):
    if nb_predict[i]==2:nb_predict[i]=3
    elif nb_predict[i]==3:nb_predict[i]=2
```

```python
print("Exatid√£o (Accuracy): {0:.4f}".format(metrics.accuracy_score(Y_treino, nb_predict)))
```

```python
nb_predict_teste=kmeans.predict(X_teste)+1
for i in range(54):
    if nb_predict_teste[i]==2:nb_predict_teste[i]=3
    elif nb_predict_teste[i]==3:nb_predict_teste[i]=2
```

```python
print("Exatid√£o (Accuracy): {0:.4f}".format(metrics.accuracy_score(Y_teste, nb_predict_teste)))
```
